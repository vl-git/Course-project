```python
text = []
article = soup.find(class_='ArticleRest')
bad_name = 'Имени АРТЕМА (АРТЕМОВО).;ВЕРХНЕЕ, ныне часть Лисичанска.;ГОСТЫНИ, ныне часть г. Плявиняс.;ДЗЕРЖИНСК (Донецкая область).;ДЗЕРЖИНСК (Минская область).;ИЫХВИ (ЙЫХВИ), ныне деревня.;КАЛИНИНСК, ныне часть Горловки.;Приказ ВГК см. г. Тильзит.;КЕМЕРИ, ныне часть Юрмалы.;КИНГИСЕПП (Ленинградская область).;КОМСОМОЛЬСК, ныне часть Горловки.;МУСТЛА, ныне деревня.;НАУЯ-ВИЛЬНЯ, ныне часть Вильнюса.;НИКИТОВКА, ныне часть Горловки.;НИКОЛАЕВ (Львовская область).;НИКОЛАЕВ (Николаевская область).;НОВЫЙ ДОНБАСС, ныне часть Снежного.;ПЕРВОМАЙСК (Ворошиловградская область).;ПЕРВОМАЙСК (Николаевская область).;ПРОЛЕТАРСК, ныне часть Лисичанска.;РОСТОВ-на-ДОНУ.;СЛОКА, ныне часть Юрмалы.;ТАЦИНСКАЯ, станица (ТАЦИНСКИЙ).'.split(';')
name = ''
buffer = ''
for chunk in article.find_all(class_='mt')[2:]:
    if len(chunk.text) == 1:
        continue
    if 'см.' in chunk.text:
        continue
    if chunk.text.isupper() or chunk.text in bad_name:
        text.append((name, buffer))
        buffer = ''
        name = chunk.text
        continue
    buffer = buffer + chunk.text
text = text[1:]
print(len(text))
print(text)
        

```

```python
source = []
exceptions = []
multiple = []
for i in text:
    if i[1].count('Оккупирован') > 1:
        multiple.append(i)
    elif i[1].startswith("Оккупирован"):
        source.append(i)
    else:
        exceptions.append(i)

print(len(source))        
print(len(multiple))
print(len(exceptions))
    
```

```python
city = []
occupated = []
rescued = []
for i in multiple:
    text = i[1].split('Оккупирован')
    for j in range(len(text)):
        if j == 0:
            continue
        try:
            lst = text[j].strip().split()[:8]
            conquer_date = lst[0] + '.' + months[lst[1]] + '.' + lst[2]
            rescue_date = lst[5] + '.' + months[lst[6]] + '.' + lst[7]
            occupated.append(conquer_date)
            rescued.append(rescue_date)
            city.append(i[0])
        except KeyError:
            exceptions.append(i)
            break
print(city)
print()
print(occupated)
print()
print(rescued)
```

```python
exceptions2 = []
for c, text in exceptions:
    if text.startswith('Городом овладели'):
        lst = text.split()[:5]
        occupated.append('-')
        rescue_date = lst[2] + '.' + months[lst[3]] + '.' + lst[4]
        rescued.append(rescue_date)
        city.append(c)
        print(c, end=' ')
        print(rescue_date)
    elif text.startswith('Освобожден'):
        lst = text.split()[:4]
        occupated.append('-')
        rescue_date = lst[1] + '.' + months[lst[2]] + '.' + lst[3]
        rescued.append(rescue_date)
        city.append(c)
        print(c, end=' ')
        print(rescue_date)
    else:
        exceptions2.append((c,text))
        
```

```python
def format_date(x):
    if x ==  '-':
        return '1941-06-22'
    if len(x) != 10:
        x = '0' + x
    day = x[:2]
    month = x[3:5]
    year = x[6:]
    return f'{year}-{month}-{day}'
```

```python
df['occupied'] = df['occupied'].apply(format_date)
df['rescued'] = df['rescued'].apply(format_date)
df
```

```python
def get_coordinates(city, bad_list):
    response = urlopen('https://ru.wikipedia.org/wiki/' + quote(city))
    if int(response.status) != 200:
        bad_list.append(city)
        return -1
    html = response.read().decode('utf8')
    soup = BeautifulSoup(html)
    a = soup.find('a', class_="mw-kartographer-maplink")
    lat = a['data-lat']
    lon = a['data-lon']
    return (lat, lon)
```

```python
bad_list = []
coordinates = {}

for city in df['city']:
    if city in coordinates:
        continue
    else:
        try:
            coordinates[city] = get_coordinates(city, bad_list)
            print(city, coordinates[city])
        except Exception:
            bad_list.append(city)
            
print(bad_list)
```
